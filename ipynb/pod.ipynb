```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# load video
import skvideo.io
import skvideo.datasets
vid_data = skvideo.io.vread('../video/c6lc01201a6.avi')
n_frames, frame_height, frame_width, n_channels = vid_data.shape
```

```python
# data matrix
X = np.empty([frame_height*frame_width*n_channels, n_frames])
for i_frame in range(n_frames):
    X[:,i_frame] = vid_data[i_frame,...].reshape(frame_height*frame_width*n_channels)
```

```python
# principal orthognal directions
def pod(X):
    XTX = np.dot(X.T,X)
    S,V = np.linalg.eig(XTX)
    S=np.sqrt(S)
    U=X.dot(V/S)
    return U,S,V

U,S,V = pod(X)
```

```python
plt.plot(S)
```

```python
fig,ax = plt.subplots(15,4,figsize=(8,20))
for i_frame in range(n_frames):
    j,k = i_frame/4, i_frame%4
    ax[j,k].imshow(S[i_frame]*U[:,i_frame].reshape(frame_height,frame_width,n_channels))
    ax[j,k].set_xlim(0,frame_width)
    ax[j,k].set_ylim(frame_height,0)
    ax[j,k].axis('off')
plt.tight_layout(h_pad=0,w_pad=0)
```
